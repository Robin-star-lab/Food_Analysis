```python
outlier_df=final_df.iloc[:,1:-1]
outlier_df
```

```python
def Outlier_detection(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR =Q3 - Q1
    lower_limit = Q1 - (1.5*IQR)
    upper_limit = Q3 + (1.5*IQR)
    
    return column.apply(lambda x: x<lower_limit or x>upper_limit)

Outliers=outlier_df.apply(Outlier_detection)

print (Outliers)
```

```python
# Droping outliers
df_cleaned=final_df[~Outliers.any(axis=1)]
df_cleaned

```

**The final dataframe is skewed to the right and has outliers.**

```python
def outlier_removal(outlier_df):
    for col in outlier_df.select_dtypes(include=['number']).columns:
        Q1 = outlier_df[col].quantile(0.25)
        Q3 = outlier_df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5*IQR
        upper_bound = Q3 + 1.5*IQR
        
        outlier_df=outlier_df[(outlier_df[col] >=lower_bound) & (outlier_df[col] <=upper_bound)]
    return outlier_df
    
outlier_free=outlier_removal(outlier_df)
outlier_free
    
```

```python
def outliers(df,ft):
    Q1 = df[ft].quantile(0.25)
    Q3 = df[ft].quantile(0.75)
    IQR =Q3 - Q1
    lower_limit=Q1 - 1.5*IQR
    upper_limit=Q3 + 1.5*IQR
    ls= df[(df[ft] < lower_limit) | (df[ft] > upper_limit)]
    return ls
outliers(food_free,"Carbohydrates")
```

```python


# Function to remove outliers using IQR for all numeric columns
def remove_outliers_iqr_all_columns(food_free):
    for col in food_free.select_dtypes(include=['number']).columns:
        Q1 = food_free[col].quantile(0.25)
        Q3 = food_free[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        food_free = food_free[(food_free[col] >= lower_bound) & (food_free[col] <= upper_bound)]
    return food_free

# Remove outliers from all numeric columns
df_no_outliers = remove_outliers_iqr_all_columns(food_free)

print(df_no_outliers)
```
